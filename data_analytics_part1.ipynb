##  F) Value Counts and Cross Tabulations

```python
# Value count function provides the occurances of each unique value in a column.|
```

```python
ipl_auction_df.COUNTRY.value_counts()
```

```python
ipl_auction_df.COUNTRY.value_counts(normalize=True)*100 
# passing parameter normalize = true to the value count function will calculate the 
# percentage of occurances of each unique value.
# this cross-tabulation features will help find occurances for the combination of values for the columns. 
```

```python
round((ipl_auction_df.COUNTRY.value_counts(normalize=True)*100),3)  # no need to write it in the copy
```

```python
## for age in a dataset: 
# 1 means the player is less than 25years
# 2 means the player is between 25 and 35 years, and
# 3 means the player is more than 35
```

```python
pd.crosstab(ipl_auction_df['AGE'],ipl_auction_df['PLAYING_ROLE'])
```

## G) Sorting data frames by Column Values

```python
# sort_values() function takes the column names, based on which the records need to be sorted.
# By default, sorting is done in ascending order.
# The following line of code selects two columns 'PLAYER NAME' and 'SOLD PRICE'
# from the DataFrame and then sorts the rows by 'SOLD PRICE'
```

```python
ipl_auction_df[['PLAYER NAME', 'SOLD PRICE']].sort_values('SOLD PRICE')[0:5]
```

```python
# To sort the dataframe in descending order, pass false to the ascending parameter. 
```

```python
ipl_auction_df[['PLAYER NAME', 'SOLD PRICE']].sort_values('SOLD PRICE',ascending=False)[0:5]
```
