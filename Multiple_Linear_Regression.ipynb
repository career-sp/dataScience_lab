```python
# date : 04/12/2024
import pandas as pd
from pandas import DataFrame
from sklearn import linear_model
import statsmodels.api as sm
```

```python
# reading the input data from csv file
df = DataFrame(pd.read_csv("stock.csv"))
df
```

```python
# here we have 2 variables for multiple regression.
X = df[['Interest_Rate', 'Unemployment_Rate']]
Y = df['Stock_Index_Price']
```

```python
# Model fitting with sklearn
regr = linear_model.LinearRegression()
regr.fit(X,Y)
```

```python
# Displaying Intercepts and co-efficients

print('Intercept value :\n',regr.intercept_)
print('Co-efficient value :\n',regr.coef_)
```

```python
# Prediction with sklearn

New_Interest_Rate = 2.75
New_Unemployment_Rate = 5.3
print('Predicted Stock Index Price: \n', regr.predict([[New_Interest_Rate, New_Unemployment_Rate]]))
```

```python
# prediction with statements.
# (a) what percentage of variation in the response is explained by these predictors (R-squared)

X = sm.add_constant(X) # adding a constant.
model = sm.OLS(Y,X).fit()
#predictions = model.predict(X)
print_model = model.summary()
print(print_model)
```
