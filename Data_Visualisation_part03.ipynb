```python
# copy

# Question: 16: Draw bar plot of a sample dataframe.

import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia', 'Japan'],
                       'Population': [1093, 1166, 1444, 1226]})

sn.barplot(x = "Population", y = "Country", data=dataset)
plt.title('Population by Country')

plt.show()
```

```python
# copy-10
import pandas as pd
ipl_auction = pd.read_csv('IPLdata.csv')
```

```python
# mean of grouped data
soldprice_by_age = ipl_auction.groupby('AGE')['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age)
```

```python
# Question - 17
# copy - 10
sn.barplot(x= 'AGE', y='SOLD PRICE', data= soldprice_by_age)
plt.title("Bar plot for average sold price versus age")

plt.xlabel('Age')
plt.ylabel('mean (SOLD PRICE)')

plt.show
```

```python
# 
# It is a plot that shows the frequency distribution of a set of continuous variables. It gives an insight into the distribution 
# of the variables, outliers, skewness etc.
# to draw a histogram we utilise hist() method of matplotlib library.
```

```python
# Histogram


plt.hist(ipl_auction['SOLD PRICE'])
plt.title("Histogram for SOLD PRICE")
plt.xlabel('SOLD PRICE')
plt.ylabel('Frequency')

plt.show()
```

```python


plt.hist(ipl_auction['SOLD PRICE'], bins=20)
plt.title("Histogram for SOLD PRICE")
plt.xlabel('SOLD PRICE')
plt.ylabel('Frequency')

plt.show()
```

```python
# Distribution or Density plot.

# It depicts the distribution of data over a continuous interval like a smooth histogram.

# copy
sn.distplot(ipl_auction['SOLD PRICE'])
plt.title("Distribution plot for SOLD PRICE")
plt.show();
```

```python
# E) Scatter Plot.

# It is useful for assessing the strength of the relationship between two variables and also find if there any outliers in 
# the dataset.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x,y)

plt.show()
```

```python

ipl_batsman_df = ipl_auction[ipl_auction['PLAYING_ROLE']=='Batsman']
plt.scatter(x = ipl_batsman_df.SIXERS, y = ipl_batsman_df['SOLD PRICE'])
plt.title("Scatter plot for SOLD PRICE versus SIXERS")

plt.ylabel('SOLD PRICE')

plt.xlabel('SIXERS')

plt.show()
```

```python
# Pair Plot.
# if there are many variables it is not convenient to draw scatter plots for each pair of variables to understand the relationship
# So a pair-plot can be use to depict the relationship in a single diagram using pairplot() method.
```

```python
influential_features = ['SR_B', 'AVE', 'SIXERS', 'SOLD PRICE']
sn.pairplot(ipl_auction[influential_features])
plt.show()
```

```python
# Correlation and Heatmap

# Correlation is used for measuring the strength and direction of the linear relationship between two continuous and random variables.
# It indicates the extends 

influential_features = ['SR_B', 'AVE', 'SIXERS', 'SOLD PRICE']
ipl_auction[influential_features].corr()

# the correlation is with itself
```

```python
sn.heatmap(ipl_auction[influential_features].corr(), annot = True)
plt.title('Heatmap of correlation values')
plt.show()
```

```python
# without using annot = True.
sn.heatmap(ipl_auction[influential_features].corr())
plt.title('Heatmap of correlation values')
plt.show()
```

```python
# Pie-plot
# a pie plot or a circle chart is a circular statistical graphic which is divided into slices to illustrate numerical proporsion.
# the area of the chart is the total percentage of the given data.

# not copy.

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
plt.pie(y)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
# not copy
y = np.array([35,25,25,15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels=mylabels)
plt.show()
```

```python
# percentage (seat share) --> autopct
import matplotlib.pyplot as plt
import numpy as np
# copy
SOLD_PRICE = [720250.000000, 484534.883721, 520178.571429]
my_labels = ['AGE 1', 'AGE 2', 'AGE 3']

plt.pie(SOLD_PRICE, labels = my_labels, autopct='%.2f%%')

plt.title('Pie plot of average sold price percentage versus age')
plt.axis('equal')
plt.legend()
plt.show()
```
