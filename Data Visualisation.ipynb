```python
#Question no 1. Draw a line in a diagram from postion (1,3) to position (8,10)

# (x1,y1) and (x2,y2)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,8])
ypoints = np.array([3,10])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
## COPY-1
#Question2: Draw a line in a diagram from the position (1,3) to (2,8) then to (6,1) and finally to position (8,10)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
#copy - 8
# the default width value is 0.8
# Draw 4 Bars

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

```python
# Horizontal bars.
# copy - 9

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.barh(x,y)
plt.show()
```

```python
# changing the color

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.barh(x,y, color='g')
plt.show()
```

```python
# height

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3,8,1,10])

plt.barh(x,y, color='g', height = 0.1)
plt.show()
```
